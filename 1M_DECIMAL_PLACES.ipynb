# Frequency of Digits in First 1 Million Decimal Places in Pi (Ï€)

```python
from mpmath import mp, pi, e
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mp.dps = 1000001
```

```python
pi_dec = str(pi).split('.')[1]
#pi_dec
```

```python
pi_dict = dict()

for i in pi_dec:
    if i in pi_dict: pi_dict[i]+=1
    else: pi_dict[i]=0

pi_dict
```

```python
pi_df = pd.DataFrame.from_dict(pi_dict, orient='index').rename(columns={0:'Repetitions'})
```

```python
plt.figure(figsize=(20,10))
plt.title('\nFREQUENCY OF DIGITS IN THE FIRST 1 MILLION DECIMAL PLACES OF Ï€\n', fontsize=30)
plt.yscale('log')
plot = sns.lineplot(data=pi_df)
plot.set_xlabel('DIGITS',fontsize=20)
plot.set_ylabel('FREQUENCY', fontsize=20)
plt.xticks(fontsize=16)
plt.legend(fontsize='xx-large')
plt.show()
```

```python
(pi_df/10000).sort_index().rename(columns={'Repetitions':'% Repetitions'})
```
